# Exploring HGCal

Let's take a look at some hgcal data.  First some simple visualizations...

```python
#imports and setup
%matplotlib inline
%cd /home/naodell/work/CMS/hgcal/hgcal_analysis''

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly as py

import scripts.helpers as hlp
```

```python
data_path = 'data/hgcalNtuple.root'
gen_features = ['genpart_energy', 'genpart_eta', 'genpart_phi', 'genpart_pid', 'genpart_dvx', 'genpart_dvy', 'genpart_dvz']
df_gen = hlp.root_to_dataframe(data_path, gen_features, index='event')

reco_features = ['rechit_energy', 'rechit_x', 'rechit_y', 'rechit_z', 'rechit_pt', 'rechit_eta', 'rechit_phi']
df_reco = hlp.root_to_dataframe(data_path, reco_features, index='event')

#skim to single event and one side (+z)
evt_gen  = df_gen.ix[3701].query('genpart_eta > 0.')
evt_reco = df_reco.ix[3701].query('rechit_eta > 0.')
```

```python
fig, axes = plt.subplots(1, 1, facecolor='white', figsize=(8,8))
axes.scatter(evt_reco.rechit_eta, evt_reco.rechit_phi, marker = 'o', s = evt_reco.rechit_energy)
axes.scatter(evt_gen.genpart_eta, evt_gen.genpart_phi, c = evt_gen.genpart_energy, s = evt_gen.genpart_energy)
axes.set_xlabel(r'$\eta$')
axes.set_ylabel(r'$\phi$')
plt.show()
```
